# Bronze Layer

## Purpose

The Bronze layer stores raw data exactly as it is received from the source systems.

### Tasks Performed

- Load raw datasets
- Add ingestion metadata
- Store data in Delta tables

### Benefits

- Data lineage
- Auditability
- Historical tracking

```python
from pyspark.sql.functions import current_timestamp

# Read raw tables

drones = spark.table("raw_drones")
deliveries = spark.table("raw_deliveries")
logs = spark.table("raw_logs")

# Add ingestion timestamp

drones_bronze = drones.withColumn(
    "ingestion_time",
    current_timestamp()
)

deliveries_bronze = deliveries.withColumn(
    "ingestion_time",
    current_timestamp()
)

logs_bronze = logs.withColumn(
    "ingestion_time",
    current_timestamp()
)

# Save Bronze tables

drones_bronze.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("bronze_drones")

deliveries_bronze.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("bronze_deliveries")

logs_bronze.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("bronze_logs")

print("Bronze layer created!")
```

```python
spark.sql("SELECT * FROM bronze_drones").show()

spark.sql("SELECT * FROM bronze_deliveries").show()

spark.sql("SELECT * FROM bronze_logs").show()
```
