# Silver Layer

## Purpose

The Silver layer transforms and cleans the raw data from the Bronze layer.

## Operations Performed

- Data cleaning
- Type conversions
- Business logic implementation
- Derived columns creation
- Data enrichment

## Derived Columns

- delivery_duration
- failure_flag
- processed_time

## Benefits

- Improved data quality
- Faster analytics
- Consistent business logic

```python
from pyspark.sql.functions import *
```

```python
drones = spark.table("bronze_drones")

deliveries = spark.table("bronze_deliveries")

logs = spark.table("bronze_logs")
```

```python
silver_deliveries = deliveries.withColumn(

    "delivery_duration",

    (
        unix_timestamp("end_time")
        -
        unix_timestamp("start_time")
    ) / 60

).withColumn(

    "processed_time",

    current_timestamp()
)
```

```python
silver_logs = logs.withColumn(

    "failure_flag",

    when(
        col("status") == "SUCCESS",
        0
    ).otherwise(1)

).withColumn(

    "processed_time",

    current_timestamp()
)
```

```python
silver_deliveries.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("silver_deliveries")

silver_logs.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("silver_logs")
```
