# Drone Delivery Monitoring & Failure Analytics Pipeline

## Objective

This notebook generates synthetic datasets for:

- Drones
- Deliveries
- Flight logs

These datasets simulate real-world drone logistics operations and will be used in the Bronze, Silver, and Gold layers of the Medallion Architecture.

## Technologies Used

- Python
- Pandas
- Databricks
- PySpark

```python
import pandas as pd
import random
from datetime import datetime, timedelta

drones = []

for i in range(1, 21):
    drones.append({
        "drone_id": i,
        "model": random.choice(
            ["DJI-X1", "SkyWing", "AeroMax", "Falcon"]
        ),
        "max_range_km": random.randint(20, 100)
    })

drones_df = pd.DataFrame(drones)

display(drones_df)
```

```python
cities = [
    "Mumbai",
    "Delhi",
    "Pune",
    "Bangalore",
    "Hyderabad"
]

deliveries = []

for i in range(1, 201):

    start = datetime.now()

    end = start + timedelta(
        minutes=random.randint(10, 120)
    )

    deliveries.append({
        "delivery_id": i,
        "drone_id": random.randint(1, 20),
        "source": random.choice(cities),
        "destination": random.choice(cities),
        "distance_km": round(
            random.uniform(2, 40),
            2
        ),
        "start_time": start,
        "end_time": end
    })

deliveries_df = pd.DataFrame(deliveries)

display(deliveries_df)
```

```python
statuses = [
    "SUCCESS",
    "BATTERY_FAILURE",
    "WEATHER_FAILURE",
    "SIGNAL_FAILURE"
]

logs = []

for i in range(1, 201):

    logs.append({
        "log_id": i,
        "delivery_id": i,
        "battery_level": random.randint(10, 100),
        "gps_signal": random.randint(30, 100),
        "status": random.choice(statuses)
    })

logs_df = pd.DataFrame(logs)

display(logs_df)
```

```python
# Convert pandas DataFrames to Spark DataFrames

drones_spark = spark.createDataFrame(drones_df)
deliveries_spark = spark.createDataFrame(deliveries_df)
logs_spark = spark.createDataFrame(logs_df)

# Save as Delta tables

drones_spark.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("raw_drones")

deliveries_spark.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("raw_deliveries")

logs_spark.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("raw_logs")

print("Raw tables created successfully!")
```
